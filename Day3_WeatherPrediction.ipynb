# Can you predict Boston's weather?
Many of you are new in town, so you may to know what the weather is going to be like in the following weeks. The dataset loaded below contains information about temperature highs, lows, and averages of temperature, dew point, wind speeds, sea level pressure, and precipitation levels for every day from 1/1/2013 - 4/8/2018 inclusive. All data is publicly available in its entirety on the Weather Underground website.

The goal, is to predict the weather in the upcoming weeks using a simple Markov chain. To do so we'll use the Pandas module: https://pandas.pydata.org

```python
import pandas as pd
import numpy as np
from random import choices

# load the dataset in csv form
df = pd.read_csv("https://raw.githubusercontent.com/RodericGuigoCorominas/datascience/main/datasets/boston_weather_2008.csv")
```
